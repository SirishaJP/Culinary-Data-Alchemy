## A/B Testing

### T-test

```python
%store -r df_a
%store -r df_b
```

```python
import pandas as pd
import scipy.stats as stats

# Perform a t-test to compare total revenue between Group A and Group B
t_stat, p_value = stats.ttest_ind(df_a['Revenue'], df_b['Revenue'])
```

```python
# Print the results
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Set the significance level (alpha)
alpha = 0.05

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis (H0). There is a significant difference in sales between Group A and Group B.")
    print("Sales are significantly higher with the new warm menu (Group B) compared to the old menu (Group A).")
else:
    print("Fail to reject the null hypothesis (H0). There is no significant difference in sales between Group A and Group B.")
```
