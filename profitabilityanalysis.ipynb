```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the menu data from a CSV file
menu_data = pd.read_csv("menudata.csv")
```

```python
menu_data
```

```python
# Calculate profit for each menu item
menu_data['Profit'] = (menu_data['SellingPrice'] - menu_data['CostPrice']) * menu_data['NumberSold']
```

```python
# Sort the menu items by profitability (descending order)
menu_data = menu_data.sort_values(by='Profit', ascending=False)
```

```python
print(menu_data.head(5))
```

```python

# Visualize profitability with a bar chart
plt.figure(figsize=(12, 6))
plt.bar(menu_data['MenuItem'], menu_data['Profit'], color='skyblue')
plt.xlabel('Menu Item')
plt.ylabel('Profit')
plt.title('Profitability Analysis of Menu Items')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the bar chart
plt.show()
```
